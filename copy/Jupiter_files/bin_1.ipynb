```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/home/aamin/files/marketing/copy/reports_data/tie_up_data.csv")
```

```python
data.drop("Unnamed: 0",axis = 1, inplace= True)
```

```python
data#.groupby(["SMS Language","Broadcast name"]).size()
```

```python
data[data["Status"] == "Delivered"].groupby(["Delivered Date","SMS Language"]).size().unstack(0)#.plot(kind = "pie", figsize= (30,30),subplots=True,autopct = "%1.1f%%")
```

```python
data[(data["Status"] != "Delivered") & (data["Clicked Hour of Day"].isna() == True)].groupby(["Delivered Date","SMS Language"]).size().unstack(0)#.plot(kind = "bar", figsize= (30,30),subplots=True)
```

```python
data.groupby(["SMS Language","Hour of Day"]).size().unstack(0).plot(kind = "pie", figsize= (30,30),subplots=True,autopct = lambda p: format(p, '.1f') if p > 2 else None)
```

```python
data.groupby(["SMS Language","Clicked Hour of Day"]).size().unstack(0).plot(kind = "pie", figsize= (30,30),subplots=True,autopct = lambda p: format(p, '.1f') if p > 2 else None)
```

```python
lang_per_hr_of_day = data[["Clicked Hour of Day","SMS Language","Day of Week"]].groupby(["SMS Language","Clicked Hour of Day"]).count() 
# lang_per_hr_of_day.loc["en",:].reset_index()

plt.figure(figsize = (20,20))
plt.subplot(3,3,1)
plt.pie(lang_per_hr_of_day.loc["en",:].reset_index()["Day of Week"], autopct = lambda p: format(p, '.1f') if p > 2 else None)
plt.legend(labels=lang_per_hr_of_day.loc["en",:].reset_index()["Clicked Hour of Day"])
plt.ylabel("English")

plt.subplot(3,3,2)
plt.pie(lang_per_hr_of_day.loc["hi",:].reset_index()["Day of Week"], autopct = lambda p: format(p, '.1f') if p > 2 else None)
plt.legend(labels=lang_per_hr_of_day.loc["hi",:].reset_index()["Clicked Hour of Day"])
plt.ylabel("Hindi")

plt.subplot(3,3,3)
plt.pie(lang_per_hr_of_day.loc["kn",:].reset_index()["Day of Week"], autopct = lambda p: format(p, '.1f') if p > 2 else None)
plt.legend(labels=lang_per_hr_of_day.loc["kn",:].reset_index()["Clicked Hour of Day"])
plt.ylabel("Kannada")

plt.show()
```

```python
lanuguage_clicked_day_wise_day = data[data["Clicked Hour of Day"].isna() == False][["Day of Week","SMS Language","Hour of Day"]].groupby(["SMS Language","Day of Week"]).count()


plt.figure(figsize = (20,20))
plt.subplot(3,3,1)
plt.pie(lanuguage_clicked_day_wise_day.loc["en",:].reset_index()["Hour of Day"], autopct = lambda p: format(p, '.1f') if p > 2 else None)
plt.legend(labels=lanuguage_clicked_day_wise_day.loc["en",:].reset_index()["Day of Week"])
plt.ylabel("English")

plt.subplot(3,3,2)
plt.pie(lanuguage_clicked_day_wise_day.loc["hi",:].reset_index()["Hour of Day"], autopct = lambda p: format(p, '.1f') if p > 2 else None)
plt.legend(labels=lanuguage_clicked_day_wise_day.loc["hi",:].reset_index()["Day of Week"])
plt.ylabel("Hindi")

plt.subplot(3,3,3)
plt.pie(lanuguage_clicked_day_wise_day.loc["kn",:].reset_index()["Hour of Day"], autopct = lambda p: format(p, '.1f') if p > 2 else None)
plt.legend(labels=lanuguage_clicked_day_wise_day.loc["kn",:].reset_index()["Day of Week"])
plt.ylabel("Kannada")

plt.show()
```
